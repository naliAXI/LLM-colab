```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Will error if it has layers on disk
checkpoint = "HuggingFaceH4/starchat-alpha"
tokenizer = AutoTokenizer.from_pretrained(checkpoint, device_map="auto", torch_dtype=torch.float16)
model = AutoModelForCausalLM.from_pretrained(checkpoint, device_map="auto", torch_dtype=torch.float16)
```

```python
input_text = "How can I write a Python function to generate the nth Fibonacci number?"
inputs = tokenizer.encode(input_text, return_tensors="pt")
outputs = model.generate(inputs, pad_token_id=tokenizer.eos_token_id, max_new_tokens= 256, temperature = 0.2, top_k=50, top_p=0.95, repetition_penalty=1.2).to('cuda')
print(tokenizer.decode(outputs[0]))
```
